# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA), is one of the most crucial steps in any data science and machine learn project. \
It involves understanding the dataset by summarizing its key characteristics, visualizing trends, identifying patterns and detect anomalies.
Performing EDA allows data scientists to make informed decisions before building machine learning models, ensuring that the dataset is clean, well-sctructure and prepared for analysis.

## Visualizing data with Matplotlib and Seaborn

```python
# Matplotlib

import matplotlib.pyplot as plt

# Sample data
years = [2015,2016,2017,2018,2019,2020]
sales = [500,700,800,950,1100,1300]

# Creating the plot
plt.plot(years, sales, marker='o', linestyle='-', color='b')
plt.xlabel('Ýear')
plt.ylabel('Sales')
plt.title('Sales growth over the years')
plt.grid()
plt.show()

```

```python
# Creating a Histogram with Seaborn 

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [22, 25, 27, 30, 34, 36, 40, 42, 45, 50]}
df = pd.DataFrame(data)

# Create the histogram
sns.histplot(df['Age'],bins=5, kde=True, color='purple')
plt.title('Age distribution of customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Identifying Patterns and Outliers 
Understanding patterns and detecting anomalies in data is crucial for making accurate predictions. Identifying trends, distributions, and unusual points (outliers) helps in improving the quality of the dataset. 

#### Finding Outliers 
Using Boxplots Outliers are values that differ significantly from the rest of the dataset. They can be due to data entry errors or actual deviations from normal trends. Identifying and handling outliers is essential because they can distort statistical analyses and machine learning models. 

```python
# Creating a Boxplot to Identify Outliers
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Salary': [35000, 42000, 50000, 52000, 60000, 75000, 90000, 110000, 150000, 250000]}
df = pd.DataFrame(data)

# Create the boxplot
sns.boxplot(y=df['Salary'])
plt.title('Boxplot of salaries')
plt.show()

# The boxplot helps us visualize the spread of salaries and identify any outliers that may exist. 
# If extreme values are observed, further investigation is needed to determine whether they should be removed or handled differently.
```

### Feature Engineering Basics 
Feature engineering is the process of transforming raw data into meaningful features that improve the predictive power of machine learning models. It involves creating new variables, selecting relevant features, and encoding categorical data to make datasets more suitable for analysis. 

#### Feature Scaling (mentioned in day 3)
Machine learning models often require features to be on a similar scale to improve convergence and performance. 
Two common methods for feature scaling are: 
- Standardization (Z-score Normalization): Converts values to have a mean of zero and a standard deviation of one.  
- Min-Max Scaling: Rescales values to fall within a fixed range, typically [0,1].

```python
# Example: Applying Feature Scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np

values = np.array([[100], [200], [300], [400], [500]])
scaler = StandardScaler()
values_standardized = scaler.fit_transform(values)
print("Standardized Values:\n", values_standardized)

min_max_scaler = MinMaxScaler()
values_min_max_scaled = min_max_scaler.fit_transform(values)
print("Min-Max Scaled Values:\n", values_min_max_scaled)
```

### Example: Creating a Correlation Heatmap to Analyze Relationships Between Features 
 
A correlation heatmap is a powerful visualization that helps understand relationships between numerical features in a dataset. 
It uses color gradients to represent correlation coefficients, making it easy to identify highly correlated variables.

```python
# Example: Generating a correlation heatmap in python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data={'Age':[22,25,27,30,34,36,40,42,45,50],
       'Salary':[35000,42000,50000,52000,60000,75000,90000,110000,150000,250000],
       'Experience':[1,3,4,5,6,8,10,12,15,20]}
df = pd.DataFrame(data)

# generating the correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# The correlation heatmap allows us to quickly identify which features are highly correlated. 
# Features with high correlation may be redundant, and removing them can help prevent multicollinearity in machine learning models.
```
