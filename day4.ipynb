# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA), is one of the most crucial steps in any data science and machine learn project. \
It involves understanding the dataset by summarizing its key characteristics, visualizing trends, identifying patterns and detect anomalies.
Performing EDA allows data scientists to make informed decisions before building machine learning models, ensuring that the dataset is clean, well-sctructure and prepared for analysis.

## Visualizing data with Matplotlib and Seaborn

```python
# Matplotlib

import matplotlib.pyplot as plt

# Sample data
years = [2015,2016,2017,2018,2019,2020]
sales = [500,700,800,950,1100,1300]

# Creating the plot
plt.plot(years, sales, marker='o', linestyle='-', color='b')
plt.xlabel('Ãear')
plt.ylabel('Sales')
plt.title('Sales growth over the years')
plt.grid()
plt.show()

```

```python
# Creating a Histogram with Seaborn 

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [22, 25, 27, 30, 34, 36, 40, 42, 45, 50]}
df = pd.DataFrame(data)

# Create the histogram
sns.histplot(df['Age'],bins=5, kde=True, color='purple')
plt.title('Age distribution of customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
