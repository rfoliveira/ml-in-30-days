# Libraries

## NumPy

NumPy is a powerful library for numerical computing. It enables efficient storage and operations on large arrays and matrices. It is widely used in ML for handling datasets, performing complex mathematical calculations, and optimizing numerical computations.

```python
import numpy as np

# Creating an array
arr = np.array([1,2,3,4,5])
print('NumPy Array:', arr)

# Performing mathematial operations
arr_squared = arr ** 2
print('Squared values', arr_squared)
```

## Pandas: Data Manipulation

```python
import pandas as pd

# Creating a Dataframe
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
}
df = pd.DataFrame(data)
print(df)
```

## Matplotlib: Data Visualization

```python
import matplotlib.pyplot as plt

# Plotting a simple graph
x = [1,2,3,4,5]
y = [10,20,25,30,40]
plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

## Scikit-Learn: Building Machine Learning Models

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample dataset
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([10, 20, 30, 40, 50])

# Creating and training the model
model = LinearRegression()
model.fit(X, y)

# Making predictions
prediction = model.predict([[6]])
print('Prediction for input 6:', prediction)
```

## Example: Writing a simple script to load and visualize a dataset

### Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load a sample dataset

```python
# Creating a simple dataset
data = {
    'year': [2015,2016,2017,2018,2019,2020],
    'sales': [500,700,800,950,1100,1300]
}
df = pd.DataFrame(data)
print(df.head())
```

### Visualizing the data 

```python
plt.plot(df['year'], df['sales'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales Growth')
plt.grid()
plt.show()
```
